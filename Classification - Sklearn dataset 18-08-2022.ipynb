```python
#Import packages
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#import dataset from sklearn
from sklearn.datasets import load_breast_cancer
bc = load_breast_cancer()
bc
```

```python
# in a dataframe
bc_df = pd.DataFrame(bc['data'],columns=bc['feature_names'])
bc_df['target'] = pd.Series(bc['target'])
bc_df.head()
```

```python
X = bc_df.drop('target', axis = 1)
y = bc_df['target']
```

```python
np.random.seed(5)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size = 0.2)
```

```python
X_train.count()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_jobs = 1)
clf.fit(X_train,y_train)
clf.score(X_train,y_train), clf.score(X_test,y_test)
```

```python
def model_evaluation(y_actual,y_preds):
    from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, classification_report
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
    
    print(f'Accuracy: {accuracy_score(y_actual,y_preds)}')
    print('')
    print(f'Precision: {precision_score(y_actual,y_preds)}')
    print('')
    print(f'Recall: {recall_score(y_actual,y_preds)}')
    print('')
    print(f'f1 Score: {f1_score(y_actual,y_preds)}')
    print('')
    print(f'r2 Score: {r2_score(y_actual,y_preds)}')
    print('')
    print(f'MAE: {mean_absolute_error(y_actual,y_preds)}')
    print('')
    print(f'MSE: {mean_squared_error(y_actual,y_preds)}')
    print('')
    print(classification_report(y_actual,y_preds))
    print('')
    

    
```

```python
y_preds = clf.predict(X_test)
model_evaluation(y_test,y_preds)
```

```python
# Another hyperparameter grid similar to rs_clf.best_params_
grid_2 = {'n_estimators': [100, 200, 300],
          'max_depth': [None, 5, 10],
          'max_features': ['auto', 'sqrt'],
          'min_samples_split': [4, 6],
          'min_samples_leaf': [1, 2]}
```

```python
from sklearn.model_selection import GridSearchCV
# Setup GridSearchCV
gs_clf = GridSearchCV(estimator=clf,
                      param_grid=grid_2,
                      cv=5, # 5-fold cross-validation
                      verbose=2) # print out progress

# Fit the RandomizedSearchCV version of clf
gs_clf.fit(X_train, y_train);
```

```python
y_preds = clf.predict(X_test)
model_evaluation(y_test,y_preds)
```
